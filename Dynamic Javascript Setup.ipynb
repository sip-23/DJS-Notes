### How to Engage with Your Coaches
- Check-In with Your Coach: Set up touchpoints with your coach to discuss your progress, challenges, and next steps.
- Prepare Your Questions: Before meeting with your coach, prepare a list of questions or topics you need help with. 
    - This preparation makes your interactions more productive and ensures you don’t forget to address any pressing concerns.
- Be Open and Honest: Share your difficulties and uncertainties as well as your successes. Open communication allows your coach to provide the most effective support.
- Act on Feedback: When your coach gives you advice or feedback, make sure to apply it.
